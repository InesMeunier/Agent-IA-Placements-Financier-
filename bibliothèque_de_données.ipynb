```python
import os
from llama_index.core import SimpleDirectoryReader, VectorStoreIndex
# from llama_index.embeddings.openai import OpenAIEmbedding
from llama_index.embeddings.huggingface import HuggingFaceEmbedding
from google.colab import userdata
# OpenAIAPIkey = userdata.get('OpenAIAPIkey')
HuggingFaceAPIkey = userdata.get('HuggingFaceAPIKey')
```

##Clé API
Me permet d'utiliser le modèle HuggingFace

```python
# os.environ["OPENAI_API_KEY"] = OpenAIAPIkey
os.environ["HUGGINGFACE_API_KEY"] = HuggingFaceAPIkey
```

##Chargement des documents

```python
documents = SimpleDirectoryReader("data").load_data()
print(f"{len(documents)} documents chargés")
```

#Création de la BDD et des vecteurs

##Transformation du texte en vecteur

```python
# embed_model = OpenAIEmbedding()
# Sélection du modèle HuggingFace opensource pour transformer le texte en vecteurs numériques
embed_model = HuggingFaceEmbedding(
    model_name = "sentence-transformers/all-MiniLM-L6-v2"
)
# Pour chaque document, le découper en morceaux, transformer chaque morceau en vecteur et stocker les vecteurs dans un index (base optimisée pour la recherche)
index = VectorStoreIndex.from_documents(
    documents,
    embed_model = embed_model
)
```

##Sauvegarde de l'index

```python
index.storage_context.persist(persist_dir="storage")
```
